```python
Parkinsons_df = Parkinsons_df.drop(labels="name", axis=1)
```

```python
Parkinsons_df.shape
```

```python
sns.pairplot(Parkinsons_df, diag_kind="kde", hue="status")
```

```python
Parkinsons_df.corr()["status"]
```

```python
Parkinsons_features_df = Parkinsons_df.drop(labels="status", axis=1)
```

```python
Parkinsons_features_df.shape
```

```python
Parkinsons_labels = Parkinsons_df["status"]
```

```python
Parkinsons_labels.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(Parkinsons_features_df, Parkinsons_labels, test_size=0.30, random_state=23)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from scipy.stats import zscore
```

```python
X_train_z = X_train.apply(zscore)
```

```python
X_train.head(5)
```

```python
X_train_z.head(5)
```

```python
X_test_z = X_test.apply(zscore)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model1 = DecisionTreeClassifier(criterion='entropy')
```

```python
model1.fit(X_train_z, y_train)
```

```python
y_predict_m1 = model1.predict(X_test_z)
```

```python
model1.score(X_train_z, y_train)
```

```python
model1.score(X_test_z, y_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_predict_m1, y_test)
```

```python
print(metrics.confusion_matrix(y_test, y_predict_m1))
```

```python
model2 = DecisionTreeClassifier(criterion='entropy', max_depth=5, min_samples_leaf=3)
```

```python
model2.fit(X_train_z, y_train)
```

```python
y_predict_m2 = model2.predict(X_test_z)
```

```python
metrics.accuracy_score(y_predict_m2, y_test)
```

```python
print(metrics.confusion_matrix(y_test, y_predict_m2))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import cross_val_score

# creating odd list of K for KNN
myList = list(range(10,100))


# empty list that will hold cv scores
cv_scores = []
n_estimators = []

# perform 10-fold cross validation
for n in myList:
    model3 = RandomForestClassifier(n_estimators=n,criterion='entropy',max_depth=3,random_state=0 )
    scores = cross_val_score(model3, X_train_z, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())
    n_estimators.append(n)


MSE = [1 - x for x in cv_scores]
min(MSE)
MSE.index(min(MSE))
best_n = myList[MSE.index(min(MSE))]
print ("The optimal number of trees is %d" % best_n)
```
