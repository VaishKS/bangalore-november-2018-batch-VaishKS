#### Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.

```python
df.dtypes
```

```python
def Goal_diff_count(df):
    df["Goal_diff_count"] = df["GoalsFor"] - df["GoalsAgainst"]
    print(df.iloc[:,[1,20]])
    a = df.sort_values(by=["Goal_diff_count"],ascending=False).head(1)
    print("Maximum\n")
    print(a["Team"])
    print("Minimum\n")
    b = df.sort_values(by=["Goal_diff_count"]).head(1)
    print(b["Team"])
```

```python
Goal_diff_count(df)
```
