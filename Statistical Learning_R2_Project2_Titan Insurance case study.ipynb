#### 1. Find the mean of old scheme and new scheme column.

```python
# Mean of old scheme
mu = sales_df['Old Scheme'].mean()
mu
```

```python
# Mean of new scheme
xbar = sales_df['New Scheme'].mean()
xbar
```

#### 2. Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs?

```python
import scipy.stats as stats
from math import sqrt
```

```python
# Setting Level of Significance as 5%
alpha = 0.05
```

```python
sd = sales_df['New Scheme'].std()
```

```python
n = 30
```

```python
se = sd / sqrt(n)
```

```python
z = (xbar - mu)/se
```

```python
pValue = 1 - stats.norm.cdf(z)
```

```python
pValue
```

```python
pValue > alpha
# pValue > apha, accept Null Hypothesis & reject the Alternate hypothesis
```

#### 3. What conclusion does the test (p-value) lead to?

```python
# Under Titan's new incentive scheme, the sales output is more
```

#### 4. Suppose it has been calculated that in order for Titan to break even, the average output must increase by Â£5000 in the scheme compared to the old scheme. If this figure is alternative hypothesis, what is: 

##### a) The probability of a type 1 error?

```python
tstat = 5000/se
tstat
```

```python
df = n-1
```

```python
p = stats.t.cdf(tstat, df=df)
p
#The calculated p-value of .86781 is the probability of committing a Type I Error
```

##### b) What is the p- value of the hypothesis test if we test for a difference of USD 5000?

```python
#H0: Difference = 0
#HA: Difference = 5000
```

```python
sd_old = sales_df['Old Scheme'].std()
se_old = (sd_old/sqrt(n))
```

```python
z_score = 5000/se_old
z_score
```

```python
p_Value = 1 - stats.norm.cdf(z_score)
p_Value
```

```python
p_Value > alpha
# pValue > apha, accept Null Hypothesis & reject the Alternate hypothesis
```

##### c) Power of the test

```python
# lower limit of the region of acceptance
ll = mu - (se_old * p_Value)
ll
```

```python
#upper limit of the region of acceptance
ul = mu + (se_old * p_Value)
ul
```

```python
# the region of acceptance is range of values between 67696.01 and 68370.66
```

```python
# The power of the test is the probability of rejecting the null hypothesis given the null hypothesis is false
# Since the region of acceptance is 67696.01 to 68370.66, the null hypothesis will be rejected 
# when the sampled avg. o/p is less than 67696.01 or greater than 68370.66.
```

```python
# Probability when the sampled avg. o/p is less than 67696.01
zleft = stats.norm.cdf((ll - mu)/sd_old)
zleft
```

```python
# Probability when the sampled avg. o/p greater than 68370.66
zright = stats.norm.cdf((ul - mu)/sd_old)
zright
```

```python
beta = zright - zleft
beta
```

```python
# Power of the test is 1-beta
1-beta
```
